```python
import pandas as pd
import matplotlib.pyplot as plt
from pywaffle import Waffle
url = "https://raw.githubusercontent.com/SergioGerman412/30DayChartChallenge/main/Day2-Neo/fallecidos_covid.csv"
datos = pd.read_csv(url, sep=";")

# Eliminar filas duplicadas basadas en la columna 'UUID' ya que una persona no puede fallecer 2 veces (dato erróneo)
datos = datos.drop_duplicates(subset=['UUID'])
```

```python
casos_por_sexo = datos['SEXO'].value_counts()
# Crear el gráfico Waffle
fig = plt.figure(
    FigureClass=Waffle,
    rows=10,
    columns=20,
    values=casos_por_sexo,
    colors=["#1f77b4", "#ff7f0e"],  # Azul para Hombres, Naranja para Mujeres
    icons=['male', 'female'],  # Iconos de hombre y mujer
    legend={'loc': 'upper left', 'bbox_to_anchor': (1, 1)},
    icon_legend=True,
    icon_size=30,
    icon_style='solid',
    figsize=(13, 8)
)

# Añadir texto en la parte inferior izquierda
plt.text(0, -0.15, "Fuente de Datos: Minsa. Autor: Sergio German", fontsize=10, color='gray', ha='left')

# Añadir conteo de MASCULINO y FEMENINO
plt.text(2.5, 0.6, f"Cantidad_Fallecidos_Masculinos: {casos_por_sexo['MASCULINO']}", fontsize=10, color='black', ha='center')
plt.text(2.49, 0.5, f"Cantidad_Fallecidos_Femeninos: {casos_por_sexo['FEMENINO']}", fontsize=10, color='black', ha='center')


# Ajustar posición del título
plt.title("Cantidad de Personas fallecidas por COVID-19 en base a su Sexo", pad=20, fontweight='bold')

plt.show()
```
