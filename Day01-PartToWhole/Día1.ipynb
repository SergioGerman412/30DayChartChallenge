```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'País': ["Rusia", "Estados Unidos", "China", "Francia", "Reino Unido", "Pakistán", "India","Corea del Norte","Israel"],
    'Ojivas_nucleares': [5889, 5244, 410, 290, 225, 170, 164, 30, 90],
    'Año_datos': [2023, 2023, 2023, 2023, 2023, 2023, 2023, 2023, 2023]
})
```

```python
data = data.sort_values(by='Ojivas_nucleares', ascending=False)

# Crear el gráfico de barras horizontal
plt.figure(figsize=(11, 9))
bars = plt.barh(data['País'], data['Ojivas_nucleares'], color='skyblue')
plt.xlabel('Número_Ojivas_Nucleares')
plt.ylabel('País')
plt.title('Países con más Ojivas Nucleares en el 2023', fontweight='bold')
plt.gca().invert_yaxis()
plt.grid(axis='x', linestyle='--', alpha=0.7)


for bar, value in zip(bars, data['Ojivas_nucleares']):
    plt.text(bar.get_width() + 10, bar.get_y() + bar.get_height()/2, f'{value}', va='center', ha='left', fontsize=10, color='black')

# Calcular y agregar el total de Ojivas_nucleares
total_ojivas = data['Ojivas_nucleares'].sum()
plt.text(0.53, 0.97, f'Total: {total_ojivas}', transform=plt.gca().transAxes, ha='right', fontsize=12)
plt.figtext(0.01, 0.01, "Autor: Sergio German. Fuente datos: Wikipedia ", ha='left', fontsize=10)
plt.tight_layout()

# Guardar el gráfico como imagen en el directorio content
plt.savefig('/content/grafico_ojivas_nucleares.png')
ruta_archivo = '/content/grafico_ojivas_nucleares.png'
print("Ruta del archivo guardado:", ruta_archivo)

# Mostrar el gráfico
plt.show()
```
